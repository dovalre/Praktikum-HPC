# Bivariate analysis


## Kolom "cement" - "strength"

### i. **Visualisasi Scatter Plot**

```python
# Buat visualiasi scatter plot pada kolom "cement" - "strength"
# CODE HERE
plt.figure(figsize=(10, 6))
sns.scatterplot(x='cement', y='strength', data=data)
plt.title('Scatter Plot of Cement vs Strength')
plt.xlabel('Cement (KG)')
plt.ylabel('Strength (MPa)')
plt.show()
```

### ii. Deskripsi

```python
# jelaskan hasil visualiasi scatter plot tersebut
# Analisis hubungan antara cement dan strength menunjukkan adanya korelasi positif, di mana kekuatan tekan beton (strength) 
# cenderung meningkat seiring bertambahnya jumlah semen (cement). Hal ini terlihat dari pola scatter plot yang menunjukkan 
# kemiringan positif, mengindikasikan bahwa peningkatan jumlah semen berpotensi menghasilkan beton dengan kekuatan tekan yang lebih tinggi. 
```

### iii. Uji Hipotesis
**Two sampled T-test** :-Uji Independent Samples t Test atau 2-sample t-test membandingkan rata-rata dua kelompok independen untuk menentukan apakah ada bukti statistik bahwa rata-rata populasi terkait berbeda secara signifikan.


Tentukan null dan alternate hypothesis.

    Secara umum, null hypothesis akan menyatakan bahwa kedua populasi yang diuji tidak memiliki perbedaan yang signifikan secara statistik.
    alternate hypothesis akan menyatakan bahwa ada perbedaan.

```python
# Buat uji hipotesis untuk menentukan null atau alternate hypothesis
# CODE HERE
median_cement = data['cement'].median()
group1 = data[data['cement'] > median_cement]['strength']
group2 = data[data['cement'] <= median_cement]['strength']

t_stat, p_value = ttest_ind(group1, group2)

print(f"Nilai t-statistic: {t_stat}")
print(f"Nilai p-value: {p_value}")

if p_value < 0.05:
    print("Kesimpulan: Null hypothesis ditolak")
else:
    print("Kesimpulan: Null hypothesis diterima")
```

### iv. Kesimpulan akhir

```python
# Berikan kesimpulan yang bisa di ambil dari analysis yg anda lakukan
# Terdapat perbedaan yang signifikan secara statistik antara dua kelompok yang diuji.
# Hal Ini membuktikan bahwa semen memiliki peranan dalam mempengaruhi kekuatan tekan beton.
```
